```python
import pandas as pd
```

```python
from datetime import date, timedelta
sdate = date(2020, 11, 5) # start date
edate = date(2020, 12, 4) # end date
delta = edate - sdate # as timedelta
for i in range(delta.days + 1):
    day = sdate + timedelta(days=i)
    print(day)
```

```python
departments = ["Dairy", "Fresh Fruit", "Fresh Vegetables","Meat", "Seafood"]
big_file = pd.DataFrame()
for department in departments:
    for i in range(delta.days + 1):
        day = sdate + timedelta(days=i)
        if str(day) != "2020-11-15":
            df = pd.read_csv("/Users/ricardolu/Desktop/To do assignment/Web Analytics/Web Crawling Project/"
                             +department+"/Whole Foods "
                             +department+" "+ str(day) +".csv")
            big_file = big_file.append(df,ignore_index = True)
```

```python
big_file
```

```python
for i,v in enumerate(big_file["Quantity Sold In"]):
    if v is None:
        big_file.loc[i,"Units"] = None
        big_file.loc[i,"Quantity"] = None
    else:
        v = str(v)
        if ('lb' in v) or ("LB" in v):
            big_file.loc[i,"Units"] = 'LBS'
            big_file.loc[i,"Quantity"] = float(v.split()[0])
        elif ('oz' in v) or ('fl oz' in v) or ("Oz" in v):
            big_file.loc[i,"Units"] = 'LBS'
            oz = float(v.split()[0])
            big_file.loc[i,"Quantity"] = oz/16
        elif 'gal' in v:
            big_file.loc[i,"Units"] = 'LBS'
            gal = float(v.split()[0])
            big_file.loc[i,"Quantity"] = gal * 8.35
        elif 'pint' in v:
            big_file.loc[i,"Units"] = 'LBS'
            pint = float(v.split()[0])
            big_file.loc[i,"Quantity"] = pint * 1.043
        elif "quart" in v:
            big_file.loc[i,"Units"] = 'LBS'
            quart = float(v.split()[0])
            big_file.loc[i,"Quantity"] = quart * 2.09
        elif ('EA' in v) or ('eggs' in v) or ('Each' in v) or ('turnip' in v):
            big_file.loc[i,"Units"] = 'EA'
            big_file.loc[i,"Quantity"] = float(v.split()[0])
        elif 'Dozen' in v:
            big_file.loc[i,"Units"] = 'EA'
            dozen = float(v.split()[0])
            big_file.loc[i,"Quantity"] = dozen * 12
        else:
            big_file.loc[i,"Units"] = None
            big_file.loc[i,"Quantity"] = None
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
big_file.to_csv("Whole Foods Master File.csv")
```
