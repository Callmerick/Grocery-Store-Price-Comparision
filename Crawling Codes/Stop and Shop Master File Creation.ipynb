# Dairy

```python
import pandas as pd
import re

#Define department and import initial file
department_1 = 'Dairy'
starting_file_1 = pd.read_csv(r'C:\Users\Home\Documents\Web Analytics\Group Project\Python Exports\Stop and Shop '+department_1+' '+start_date+'.csv')

#loop over all dates in date_list to import and append files to the starting_file
for i in date_list:  
    dairy_temp = pd.read_csv(r'C:\Users\Home\Documents\Web Analytics\Group Project\Python Exports\Stop and Shop '+department_1+' '+i+'.csv')
    starting_file_1 = starting_file_1.append(dairy_temp)
    
#rename the resultant large file and reset the index
dairy_master = starting_file_1.reset_index(drop = True)

#Extract numerical values from "Quantity Sold In" and store in their own list for later inclusion in the "Quantity" column
numbers_list_1 = []
for i in range(len(dairy_master['Store'])):
    if dairy_master['Quantity Sold In'][i].split()[0] == 'APX':
        numbers_list_1.append(dairy_master['Quantity Sold In'][i].split()[1])
    else:
        numbers_list_1.append(dairy_master['Quantity Sold In'][i].split()[0])
        
for i in range(len(numbers_list_1)):
    if numbers_list_1[i] == '1/2':
        numbers_list_1[i] = '.5'
        
#Extract "units" values from "Quantity Sold In" and store in their own list for later inclusion in the "Units" column
units_list_1 = []
for i in range(len(dairy_master['Store'])):
    if dairy_master['Quantity Sold In'][i].split()[1] == 'DOZ':
        units_list_1.append('EA')
    elif dairy_master['Quantity Sold In'][i].split()[0] == 'APX':
        units_list_1.append(dairy_master['Quantity Sold In'][i].split()[2])    
    else:
        units_list_1.append(dairy_master['Quantity Sold In'][i].split()[1])
            
#Create two new columns. One to hold the Quantity number and the other to hold the Units
dairy_master['Quantity'] = numbers_list_1
dairy_master['Units'] = units_list_1         

```

# The Making of the Overall Master

```python
import pandas as pd
import re

#Now that we have five master files, one for each department, we must put them together into one Stop and Shop Master
Stop_and_Shop_Master = dairy_master.append(seafood_master).append(meat_master).append(fresh_vegetables_master).append(fresh_fruit_master)

#rename the resultant large file and reset the index
Stop_and_Shop_Master = Stop_and_Shop_Master.reset_index(drop = True)

#Cleaning the "Quantity" column
cleaned_quantity = []
for i in Stop_and_Shop_Master['Quantity']:
    if '-' in i:
        cleaned_quantity.append(i[0])
    else:
        cleaned_quantity.append(i)
Stop_and_Shop_Master['Quantity'] = cleaned_quantity
Stop_and_Shop_Master['Quantity'] = Stop_and_Shop_Master['Quantity'].astype(float)

#Cleaning the "Units" column
cleaned_units = []
for i in Stop_and_Shop_Master['Units']:
    if i == '-':
        cleaned_units.append('OZ')
    elif i == 'CT':
        cleaned_units.append('EA')
    elif i == 'GAL':
        cleaned_units.append('GALLON')
    else:
        cleaned_units.append(i)
Stop_and_Shop_Master['Units'] = cleaned_units

#Cleaning the "Price" column
Stop_and_Shop_Master['Price'] = Stop_and_Shop_Master['Price'].astype(float)

#Convert all values in the "Quantity" column to their "Pound Equivalents" (except "EA")
Stop_and_Shop_Master_2 = Stop_and_Shop_Master.copy()

gallons_to_pounds = 8.35
liters_to_pounds = 2.205
ounces_to_pounds = .0625
pints_to_pounds = 1.043
quarts_to_pounds = 2.09

converted_quantity = []
for i in range(len(Stop_and_Shop_Master_2['Quantity'])):
    if Stop_and_Shop_Master_2['Units'][i] == 'GALLON':
        converted_quantity.append(Stop_and_Shop_Master_2['Quantity'][i] * gallons_to_pounds)
    elif Stop_and_Shop_Master_2['Units'][i] == 'LTR':
        converted_quantity.append(Stop_and_Shop_Master_2['Quantity'][i] * liters_to_pounds)
    elif Stop_and_Shop_Master_2['Units'][i] == 'OZ':
        converted_quantity.append(Stop_and_Shop_Master_2['Quantity'][i] * ounces_to_pounds)
    elif Stop_and_Shop_Master_2['Units'][i] == 'PINT':
        converted_quantity.append(Stop_and_Shop_Master_2['Quantity'][i] * pints_to_pounds)
    elif Stop_and_Shop_Master_2['Units'][i] == 'QUART':
        converted_quantity.append(Stop_and_Shop_Master_2['Quantity'][i] * quarts_to_pounds)
    else:
        converted_quantity.append(Stop_and_Shop_Master_2['Quantity'][i])
    
Stop_and_Shop_Master_2['Quantity'] = converted_quantity
               
#To reflect the above standardization of the "Quantities" column into pound measurements, convert all text values in the
#"Units" column to be "LBS", unless they read "EA"
converted_units = []
for i in Stop_and_Shop_Master_2['Units']:
    if i != 'EA':
        converted_units.append('LBS')
    else:
        converted_units.append(i)

Stop_and_Shop_Master_2['Units'] = converted_units

#Outsheet final file to csv
Stop_and_Shop_Master_2.to_csv(r'C:\Users\Home\Documents\Web Analytics\Group Project\Python Exports\Stop and Shop Master File.csv',index=False)
```
