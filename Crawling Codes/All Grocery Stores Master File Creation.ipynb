# Prep Whole Foods Master

```python
import pandas as pd
import numpy as np

#Get the columns in the Whole Foods file right order
whole_foods_df = pd.read_csv(r'C:\Users\Home\Documents\Web Analytics\Group Project\Whole Foods Master File.csv')
whole_foods_df.drop('Unnamed: 0',axis = 1, inplace=True)
units_column = whole_foods_df.pop('Units')
whole_foods_df['Units'] = units_column

#Deal with the NaNs in the Quantity and Units columns and place "Beef" and "Chicken" into the "Meat" category in the Whole Foods file
quantity = []
units = []
product = []

for i in range(len(whole_foods_df['Store'])):
    if pd.isna(whole_foods_df['Units'][i]) == True:
        units.append('EA')
    else:
        units.append(whole_foods_df['Units'][i])

for i in range(len(whole_foods_df['Store'])):
    if pd.isna(whole_foods_df['Quantity'][i]) == True:
        quantity.append(1)
    else:
        quantity.append(whole_foods_df['Quantity'][i])
        
for i in range(len(whole_foods_df['Store'])):
    if whole_foods_df['Product Type'][i] in ['Chicken','Beef']:
        product.append('Meat')
    else:
        product.append(whole_foods_df['Product Type'][i])
        
#Convert dates to "yyyy-mm-dd" format
dates = []
for i in whole_foods_df['Date']:
    if len(i.split('/')) > 2:
        d = i.split('/')
        d_year = d[2]+'20'
        d_month = d[0]

        if len(d[1]) == 1:
            d_day = '0'+d[1]
        else:
            d_day = d[1]
        d_reformatted = d_year+'-'+d_month+'-'+d_day
        dates.append(d_reformatted)
    else: dates.append(i)

#Add updated columns to make final dataframe
whole_foods_df2 = whole_foods_df.copy()
whole_foods_df2['Date'] = dates
whole_foods_df2['Quantity'] = quantity
whole_foods_df2['Units'] = units
whole_foods_df2['Product Type'] = product
```
